Wir wählen 6660hz als Abtastfrequenz.
Die Niquist-Frequenz beträgt demnach 3330hz.

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
import redlab as rl
```

```python
frequenz = 6660
nyquist_frequemz = 3330
n = []
for i in range(0,7):
    n.append(int((frequenz / 4) + np.ceil(i * 832.5)))
```

```python
for i in range(0,7):
    while (input() != "q"):
        pass
    messung = rl.cbVInScan(0,0,0,1000,frequenz, 1)
    plt.plot(messung)
    plt.xlabel("Abtastpunkt (1000/s)")
    plt.ylabel("Spannung in Volt")
    plt.savefig(f'{n[i]}.png')
```
